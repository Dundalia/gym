```python
# Load all sheets with dynamic header detection
excel_file = pd.ExcelFile('db.xslx')
print(f"Available sheets: {excel_file.sheet_names}\n")

sheets_dict = {}
for sheet_name in excel_file.sheet_names:
    # First, read without header to find the correct row
    temp_df = pd.read_excel('db.xslx', sheet_name=sheet_name, header=None)
    
    # Find the row where first column is "Gruppo muscolare"
    header_row = None
    for idx, value in enumerate(temp_df[0]):
        if value == "Gruppo muscolare":
            header_row = idx
            break
    
    # Now read with the correct header
    if header_row is not None:
        sheets_dict[sheet_name] = pd.read_excel('db.xslx', sheet_name=sheet_name, header=header_row)
        print(f"Sheet: {sheet_name} (header found at row {header_row})")
        print(f"  Shape: {sheets_dict[sheet_name].shape}")
        print(f"  Columns: {list(sheets_dict[sheet_name].columns[:5])}")
    else:
        print(f"Sheet: {sheet_name} - No header row found with 'Gruppo muscolare'")
    print()
```

```python
# Aggregate all sheets into one dataframe
df_list = []
for sheet_name, sheet_df in sheets_dict.items():
    # Add the "Classe" column with the sheet name
    sheet_df['Classe'] = sheet_name
    df_list.append(sheet_df)

# Combine all dataframes
df_combined = pd.concat(df_list, ignore_index=True)

print(f"Combined dataframe shape: {df_combined.shape}")
print(f"\nClasse distribution:")
print(df_combined['Classe'].value_counts())
print(f"\nFirst few rows:")
df_combined.head()
```

```python
# Explore all columns
print("All columns in the dataframe:")
print("="*80)
for i, col in enumerate(df_combined.columns, 1):
    print(f"{i}. {col}")
print(f"\nTotal columns: {len(df_combined.columns)}")
```

```python
# Analyze key columns for filtering (columns with limited unique values)
key_columns = ['Gruppo muscolare', 'Attrezzo', 'Livello', 'Obiettivo', 'Classe']

print("Potential filter columns analysis:")
print("="*80)
for col in key_columns:
    if col in df_combined.columns:
        unique_count = df_combined[col].nunique()
        print(f"\n{col}:")
        print(f"  Unique values: {unique_count}")
        print(f"  Values: {sorted(df_combined[col].dropna().unique().tolist())[:10]}")
        if unique_count <= 30:
            print(f"  Counts:")
            print(df_combined[col].value_counts().head(10).to_string())
```

```python
# Identify all columns suitable for filters (less than 50 unique values)
print("Columns suitable for filtering (<50 unique values):")
print("="*80)
filter_candidates = []

for col in df_combined.columns:
    unique_count = df_combined[col].nunique()
    if unique_count < 50 and unique_count > 1:
        filter_candidates.append((col, unique_count))
        
# Sort by number of unique values
filter_candidates.sort(key=lambda x: x[1])

for col, count in filter_candidates:
    print(f"{col}: {count} unique values")
```

```python
# Summary of data quality
print("Data quality summary:")
print("="*80)
print(f"Total rows: {len(df_combined)}")
print(f"\nMissing values per column:")
missing = df_combined.isnull().sum()
missing_pct = (missing / len(df_combined) * 100).round(2)
missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_pct
})
print(missing_df[missing_df['Missing Count'] > 0].sort_values('Missing Count', ascending=False).head(15))
```

```python
# Fill missing "Attrezzo" values for Calisthenics with "Corpo Libero"
print("Before filling:")
print(f"Missing Attrezzo values: {df_combined['Attrezzo'].isnull().sum()}")
print(f"Missing in Chalistenics: {df_combined[df_combined['Classe'] == 'Chalistenics']['Attrezzo'].isnull().sum()}")

df_combined.loc[df_combined['Classe'] == 'Chalistenics', 'Attrezzo'] = df_combined.loc[df_combined['Classe'] == 'Chalistenics', 'Attrezzo'].fillna('Corpo Libero')

print(f"\nAfter filling:")
print(f"Missing Attrezzo values: {df_combined['Attrezzo'].isnull().sum()}")
print(f"Missing in Chalistenics: {df_combined[df_combined['Classe'] == 'Chalistenics']['Attrezzo'].isnull().sum()}")

print(f"\nAttrezzo value counts for Chalistenics:")
print(df_combined[df_combined['Classe'] == 'Chalistenics']['Attrezzo'].value_counts().head(10))
```
